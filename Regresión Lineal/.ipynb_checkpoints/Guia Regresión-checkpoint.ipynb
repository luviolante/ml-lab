# Guia regresión 
## Consigna
Para el dataset a continuación, calcular la regresión polinómica e ir variando el orden hasta obtener el resultado óptimo de AIC y BIC. También calcular $ R^2 $ y los p values.

```python
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd

X = np.load('X.npy')
y = np.load('y.npy')
print(X.size)

plt.figure()
plt.plot(X, y,'.',  color='black')
plt.title('Dataset')
plt.show()
```

## Solución con Regresión Polinómica

```python
#Separo la muestra en datos de entrenamiento y datos de prueba

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
#print(X_train.size)

#Statsmodels 
#Entreno el modelo de orden 1
lm = smf.OLS(y_train,X_train).fit()

#Calculo p-values y descarto los coeficientes mayores a 0,05
print("PValues:",lm.pvalues,"\n") 

#Calculo AIC y BIC. Miro BIC
print(lm.summary())

#Grafico
plt.figure(1)
plt.plot(X, y,'.',  color='black')
plt.title('Complete Dataset')

plt.figure(2)
plt.title('Training Subset')
plt.plot(X_train, y_train,'.',  color='blue')
plt.plot(X_train, lm.predict(X_train), c='red', linewidth=2)
plt.show()
```

```python
#Genero nueva variable

N = 3
poly_2=list()
for i in X_train:
    poly_2.append([i])
    for j in range(2,N):
        poly_2[-1].append(i**j)
        
#print(new_x)

#Entreno el modelo de orden 2
lm2 = smf.OLS(y_train, poly_2).fit()

#Calculo p-values y descarto los coeficientes mayores a 0,05
print("PValues:",lm2.pvalues,"\n") 

#Calculo AIC y BIC. Miro BIC
print(lm2.summary())

#Grafico
plt.figure(1)
plt.plot(X, y,'.',  color='black')
plt.title('Complete Dataset')

plt.figure(2)
plt.title('Training Subset')
plt.plot(X_train, y_train,'.',  color='blue')
plt.plot(X_train, lm2.predict(poly_2),'.' ,c='red', linewidth=2)
plt.show()
```

```python
#Genero nueva variable
N = 4
poly_3=list()
for i in X_train:
    poly_3.append([i])
    for j in range(2,N):
        poly_3[-1].append(i**j)

#Entreno el modelo de orden 3
lm3 = smf.OLS(y_train, poly_3).fit()

#Calculo p-values y descarto los coeficientes mayores a 0,05
print("PValues:",lm3.pvalues,"\n") 

#Calculo AIC y BIC. Miro BIC
print(lm3.summary())

#Grafico
plt.figure(1)
plt.plot(X, y,'.',  color='black')
plt.title('Complete Dataset')

plt.figure(2)
plt.title('Training Subset')
plt.plot(X_train, y_train,'.',  color='blue')
plt.plot(X_train, lm3.predict(poly_3),'.' ,c='red', linewidth=2)
plt.show()
```

```python
#Genero nueva variable
N = 5
poly_4=list()
for i in X_train:
    poly_4.append([i])
    for j in range(2,N):
        poly_4[-1].append(i**j)
        
#print(new_x)

#Entreno el modelo de orden 4
lm4 = smf.OLS(y_train, poly_4).fit()

#Calculo p-values y descarto los coeficientes mayores a 0,05
print("PValues:",lm4.pvalues,"\n") 

#Calculo AIC y BIC. Miro BIC
print(lm4.summary())

#Grafico
plt.figure(1)
plt.plot(X, y,'.',  color='black')
plt.title('Complete Dataset')

plt.figure(2)
plt.title('Training Subset')
plt.plot(X_train, y_train,'.',  color='blue')
plt.plot(X_train, lm4.predict(poly_4),'.' ,c='red', linewidth=2)
plt.show()
```

# Resultados

Mirando el BIC de los modelos resultantes, notamos que se ajusta mejor el del polinomio de orden 3. Por otro lado, esto es consistente con el hecho de que en el modelo con polinomio de orden 4, el p-value del coeficiente que acompana el grado 4 es mayor al 0,05 y por ende no es significativo para el mismo, sugiriendo una polinomial de un orden menor.

```python
#Grafico de datos de prueba con modelo de orden 3
N = 4
test_poly_3=list()
for i in X_test:
    test_poly_3.append([i])
    for j in range(2,N):
        test_poly_3[-1].append(i**j)

plt.figure(1)
plt.plot(X, y,'.',  color='black')
plt.title('Complete Dataset')

plt.figure(2)
plt.title('Testing Subset')
plt.plot(X_train, y_train,'.',  color='blue')
plt.plot(X_train, lm3.predict(poly_3),'.' ,c='red', linewidth=2)
plt.plot(X_test, lm3.predict(test_poly_3),'.' ,c='green', linewidth=2)
plt.show()
```
